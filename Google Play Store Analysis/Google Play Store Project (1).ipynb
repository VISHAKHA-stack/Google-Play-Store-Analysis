## Data Cleaning

#### Google Play Store Dataset

```python
data.isnull().sum()
```

```python
data['Category'].unique()
```

```python
data['Rating'].unique()
```

Here the dtype is object but to do operations we need int or float so lets convert the dtype to float for convenience. Then we'll come back to the Category issue.

```python
data['Rating'] = pd.to_numeric(data['Rating'])
data['Rating'].dtype
```

```python
data[data['Category']=='1.9']
```

```python
#shift 
#fill nan value with lifestyle REASON:--->
#1)Open google playstore 2)Search for this particular app 3) Look at the section mentioned in it
replacing=data[data['Rating']==19].iloc[:,1:].shift(1,fill_value='LIFESTYLE',axis=1)
replacing
```

```python
data.loc[[10472],1:]=np.array(replacing)
```

```python
#converting the only string element to float
data['Rating']=data['Rating'].replace(['1.9'],1.9)
```

```python
data['Reviews'].unique()
```

```python
data['Reviews'] = pd.to_numeric(data['Reviews'])
data['Reviews'].dtype
```

```python
data['Size'].unique()
```

```python
data['Size'] = data['Size'].replace('Varies with device', np.nan)
data['Size'] = data['Size'].str.replace('M', '000')
data['Size'] = data['Size'].str.replace('k', '')
data['Size'] = data['Size'].str.replace('1,000+', '1000')
data['Size']= data['Size'].astype(float)
data['Size'].dtype
```

```python
data['Installs'].unique()
```

```python
data['Installs'] = data['Installs'].str.replace(',', '')
data['Installs'] = data['Installs'].str.replace('+', '')
data['Installs'] = data['Installs'].astype(float)
data['Installs'].dtype
```

```python
data['Price'].unique()
```

```python
data['Price'] = data.Price.replace('Everyone', np.nan)
data['Price'] = data.Price.str.replace('$', '')
data['Price'] = data['Price'].astype(float)
data['Price'].dtype
```

```python
data['Last Updated'].unique()
```

```python
l=[]
import re
for x in data['Last Updated']:
    l.append(re.split(' |,',x))

last_updated=pd.DataFrame(l).drop(2,axis=1).rename(columns={0:'Last_Updated_Month',1:'Last_Updated_Day',3:'Last_Updated_Year'})
```

```python
data=pd.concat([data,last_updated],axis=1)
```

```python
data.loc[pd.isna(data['Current Ver'])]
```

```python
data['Current Ver'].mode()
```

```python
data['Current Ver'].fillna('Varies with device',inplace=True)
```

```python
data['Android Ver'].mode()
```

```python
data['Android Ver'].fillna('4.1 and up',inplace=True)
```

```python
data['Current_Ver_Varies_with_device']=data['Current Ver'].apply(lambda x:1 if x=='Varies with device' else 0)
data['Android_Ver_Varies_with_device']=data['Android Ver'].apply(lambda x:1 if x=='Varies with device' else 0)
```

#### What are the Top 10 installed apps in any category?

```python
def findtop10incategory(str):
    str = str.upper()
    top10 = data[data['Category'] == str]
    top10apps = top10.sort_values(by='Installs', ascending=False).head(10)
    # Top_Apps_in_art_and_design
    plt.figure(figsize=(15,12))
    plt.title('Top 10 Installed Apps',size = 20);    
    graph = sns.barplot(x = top10apps.App, y = top10apps.Installs)
    graph.set_xticklabels(graph.get_xticklabels(), rotation= 45, horizontalalignment='right');
    
    
findtop10incategory('Sports')
```

From the above graph, we can see that in the Sports category FIFA Soccer, and Dream League Soccer 2018 has the highest installs. In the same way by passing different category names to the function, we can get the top 10 installed apps.

#### What is the final distribution of Ratings for the Apps?

```python
# plot the graphs of reviews, size, installs and price per rating
rating_df = data.groupby('Rating').sum().reset_index()

fig, axes = plt.subplots(1, 4, figsize=(14, 4))

axes[0].plot(rating_df['Rating'], rating_df['Reviews'], 'r')
axes[0].set_xlabel('Rating')
axes[0].set_ylabel('Reviews')
axes[0].set_title('Reviews Per Rating')

axes[1].plot(rating_df['Rating'], rating_df['Size'], 'g')
axes[1].set_xlabel('Rating')
axes[1].set_ylabel('Size')
axes[1].set_title('Size Per Rating')

axes[2].plot(rating_df['Rating'], rating_df['Installs'], 'g')
axes[2].set_xlabel('Rating')
axes[2].set_ylabel('Installs (e+10)')
axes[2].set_title('Installs Per Rating')

axes[3].plot(rating_df['Rating'], rating_df['Price'], 'k')
axes[3].set_xlabel('Rating')
axes[3].set_ylabel('Price')
axes[3].set_title('Price Per Rating')

plt.tight_layout(pad=2)
plt.show()
```

From the above plottings, we can imply that most of the apps with higher rating range of 4.0 - 4.7 are having high amount of reviews, size, and installs. In terms of price, it doesn't reflect a direct relationship with rating, as we could see a fluctuation in term of pricing even at the range of high rating.
